```python
import os
import shutil

# Define the directory where the files are located
input_dir = '/kaggle/input/include'

# Find the categories automatically
categories = {}
for root, dirs, files in os.walk(input_dir):
    for dir in dirs:
        # Extract the category name from the folder name
        try:
            category = dir.split('. ')[1].lower()
            categories[category] = os.path.join(root, dir)
        except IndexError:
            # Skip folders that don't have a category name
            pass

# Create the output directory
output_dir = 'output'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Process each category
for category, path in categories.items():
    # Create the category folder
    category_dir = os.path.join(output_dir, category)
    if not os.path.exists(category_dir):
        os.makedirs(category_dir)

    # Find the first file in the category
    for root, dirs, files in os.walk(path):
        if files:
            first_file = files[0]
            break
    else:
        continue

    # Copy the first file to the category folder
    src_path = os.path.join(root, first_file)
    dst_path = os.path.join(category_dir, first_file)
    shutil.copy2(src_path, dst_path)
    print(f'Copied {first_file} to {category_dir}')
```

```python
import os

def find_file_path(file_name, input_dir='/kaggle/input/include'):
    """
    Finds the path of the given file in the input directory.
    
    Parameters:
    file_name (str): The name of the file to search for.
    input_dir (str): The directory to search in. Default is '/kaggle/input/include'.
    
    Returns:
    str: The full path of the file, if found. None if not found.
    """
    for root, dirs, files in os.walk(input_dir):
        if file_name in files:
            return os.path.join(root, file_name)
    return None

# Example usage
input_file = "MVI_5302.MOV"
file_path = find_file_path(input_file)
if file_path:
    print(f"Found {input_file} at: {file_path}")
else:
    print(f"{input_file} not found in the input directory.")
```

```python
import os
import cv2
import numpy as np
import mediapipe as mp

# Define the input and output directories
output_dir = '/kaggle/working/output'
skeleton_dir = '/kaggle/working/skeleton'

# Initialize MediaPipe Holistic model
mp_holistic = mp.solutions.holistic
holistic = mp_holistic.Holistic(static_image_mode=False, min_detection_confidence=0.5, min_tracking_confidence=0.5)

# Iterate through the output directory and process each video
for category in os.listdir(output_dir):
    category_dir = os.path.join(output_dir, category)
    skeleton_category_dir = os.path.join(skeleton_dir, category)
    if not os.path.exists(skeleton_category_dir):
        os.makedirs(skeleton_category_dir)

    for file_name in os.listdir(category_dir):
        input_file_path = os.path.join(category_dir, file_name)
        output_file_path = os.path.join(skeleton_category_dir, file_name)

        # Initialize video capture
        cap = cv2.VideoCapture(input_file_path)

        # Get video properties
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        fps = int(cap.get(cv2.CAP_PROP_FPS))

        # Initialize video writer
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        out = cv2.VideoWriter(output_file_path, fourcc, fps, (width, height))

        while cap.isOpened():
            # Read a frame from the video
            ret, frame = cap.read()
            if not ret:
                break

            # Convert the frame to RGB
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

            # Process the frame with MediaPipe Holistic
            results = holistic.process(frame_rgb)

            # Create a blank frame for drawing keypoints
            keypoints_frame = np.zeros((height, width, 3), dtype=np.uint8)

            # Convert landmarks to pixel coordinates and draw on the keypoints frame
            if results.pose_landmarks:
                for landmark in results.pose_landmarks.landmark:
                    x = int(landmark.x * width)
                    y = int(landmark.y * height)
                    cv2.circle(keypoints_frame, (x, y), 5, (0, 255, 0), -1)

            if results.face_landmarks:
                for landmark in results.face_landmarks.landmark:
                    x = int(landmark.x * width)
                    y = int(landmark.y * height)
                    cv2.circle(keypoints_frame, (x, y), 2, (255, 0, 0), -1)

            if results.left_hand_landmarks:
                for landmark in results.left_hand_landmarks.landmark:
                    x = int(landmark.x * width)
                    y = int(landmark.y * height)
                    cv2.circle(keypoints_frame, (x, y), 3, (0, 0, 255), -1)

            if results.right_hand_landmarks:
                for landmark in results.right_hand_landmarks.landmark:
                    x = int(landmark.x * width)
                    y = int(landmark.y * height)
                    cv2.circle(keypoints_frame, (x, y), 3, (0, 0, 255), -1)

            # Write the keypoints frame to the output video
            out.write(keypoints_frame)

            
        # Release the video capture and writer
        cap.release()
        out.release()

# Close all windows
cv2.destroyAllWindows()
```
